**Common Terminologies**

1. **Plot:** It is an illustration that can be represented as graph
2. **Figure:**  It is a diagram or a shape that can be formed by collection of plots in different dimensions.
3. **Label:** It is used to add labels or names to respective x and y axes
4. **Title:** It is used to display the title of the graph
5. **Grid:** It is a collction of objects and functions which is concerned with 3-D data

# **Problem statement**

Consider that a survey has to be done on how much distance the vehicles have covered in a span of five days.

Plot the data using different plots

```python
import pandas as pd

vehiclesData = pd.DataFrame({
    'ENFIELD': [50,40,70,80,20],
    'HONDA': [80,20,20,50,60],
    'YAMAHA': [70, 20,60,40,60],
    'KTM': [80,20,20,50,60]
}, index= ['DAY1','DAY2','DAY3','DAY4','DAY5'])
vehiclesData
```

```python
#lets plot line plot for the above data
import matplotlib.pyplot as plt
%matplotlib inline

x = [1,2,3,4,5]
enf = [50,40,70,80,20]
hnd = [60,30,20,70,70]
ymh = [70, 20,60,40,60]
ktm = [80,20,20,50,60]

#lets plot the line plot
plt.plot(x,enf,'g',label='Enfield',linewidth=5)
plt.plot(x,hnd,'c',label='Honda',linewidth=5)
plt.plot(x,ymh,'k',label='Yamaha',linewidth=5)
plt.plot(x,ktm,'y',label='KTM',linewidth=5)

plt.title('Bike Details')
plt.xlabel('Days')
plt.ylabel('Distace in kms')
plt.legend()
plt.show()
```

# **Bar Chart Plot**

Categorical data can be represented in rectangular blocks with different heights or lengths proportional to the values.

```python
#bar plot
import matplotlib.pyplot as plt

plt.bar([0.25, 1.25, 2.25, 3.25,4.25],enf,
        label="Enfield",color = 'c', width=0.5)

plt.bar([0.25, 1.25, 2.25, 3.25,4.25],hnd,
        label="Honda",color='r', width=0.5)

plt.bar([0.25, 1.25, 2.25, 3.25,4.25],ymh,
        label="Yamaha", color = 'y',width=0.5)

plt.bar([0.25, 1.25, 2.25, 3.25,4.25],ktm,
        label="KTM", color = 'g',width=0.5)

plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance(kms)')
plt.title('Bikes Details in Bar Plotting')
plt.show()
```

# **Area Plot**

This type of plotting is basically used quantitative data. A line chart forms the basis of an area plot, where the region between the axis and the line is represented by colors.  

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
#using the values of of vehicals from the previous
#cell
plt.plot([],[],color='k', label='Enfield', linewidth=5)
plt.plot([],[],color='c', label='Honda', linewidth=5)
plt.plot([],[],color='y', label='Yamaha', linewidth=5)
plt.plot([],[],color='m', label='KTM', linewidth=5)

#lets use stack plots
plt.stackplot(days, enf,hnd,ymh,ktm, colors=['k','c','y','m'])
plt.xlabel('Days')
plt.ylabel('Distance in kms')
plt.title('Bike Details in area plot')
plt.legend()
plt.show()
```

# **PIE Plot**

In this type of plots, the statistical data can represented in a circular graph where the circle is divided into portions that denote a particular data, that is , each portion is called as slice, which is proportional to different values in the data.
This sort of plot can be mainly used in mass media and business.

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

slices = [8,5,5,6]
activities = ['Enfield','Honda','Yamaha','KTM']
cols = ['c','g','y','b']
plt.pie(slices,labels = activities, colors=cols,
        startangle=90, shadow=True,
        explode=(0,0.1,0,0), autopct='%1.1f%%')
plt.title('Bike details in Pie Plot')
```

# **Scatter Plot**

Its a dot based ploting of multiple variables along x and y axes.

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

#taking previous values of bikes
plt.scatter(days, enf, label='Enfield', color = 'r')
plt.scatter(days, hnd, label='Honda', color = 'b')
plt.scatter(days, ymh, label='Yamaha', color = 'y')
plt.scatter(days, ktm, label='KTM', color = 'k')

plt.xlabel('Days')
plt.ylabel('Distance in kms')
plt.title('Bike Details in Scatter Plot')
plt.legend()
plt.show()
```

# **3D Plot**

Plotting of data along x,y,z axes to enhance the display of data represents the 3D plotting. 3D plotting is an advanced plotting technique that gives us a better view of the data representation along the three axes of the graph.

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
x = [1,2,3,4,5]
#taking the vehicle values from prev data
plt.plot(x,enf,'g',label='Enfield',linewidth=5)
plt.plot(x,hnd,'c',label='Honda',linewidth=5)
plt.plot(x,ymh,'k',label='Yamaha',linewidth=5)
plt.plot(x,ktm,'y',label='KTM',linewidth=5)

plt.xlabel('Days')
plt.ylabel('Distance in kms')
plt.title('Bike Details in 3D Plot')
plt.legend()
plt.show()
```

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt


fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x + y
ax.scatter(x,y,z, c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```

# **Contour Plots**

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

def fn(x,y):
  return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)

X, Y = np.meshgrid(x,y)
Z = fn(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Contour Plot')
plt.show()
```
