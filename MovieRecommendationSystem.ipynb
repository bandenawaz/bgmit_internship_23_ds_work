**Task:**
Print me all the movies
1. That were released between 2005 and 2010
2. have ratings above 8.0
3. Which made below the 25th percentile in revenue(hint: use quanitle method, quantile(0.25))

```python
movies_df[
    ((movies_df['year'] >= 2005) & (movies_df['year'] <= 2010))
    &(movies_df['rating'] > 8.0)
    &(movies_df['revenue'] < movies_df['revenue'].quantile(0.25))
]
```

**Applying Functions**


```python
#lets write a function to seggregate the movies into
#good movies, average movies, below average movies
def rating_function(x):
  """Function to check the rating of a movie
  and recommend accordingly based on rating"""
  if(x >= 8.0):
    return "Must Watch"
  elif(x >= 6.0 and x < 8.0):
    return "Good Movie"
  else:
    return "Watch at your own risk"
```

```python
movies_df['recomending'] = movies_df['rating'].apply(rating_function)
movies_df.head()
```
